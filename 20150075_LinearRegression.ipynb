# IMPORT THE NECESSARY PACKAGES
Pandas is used for import data with *.cvs format

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Split the data into training/testing sets

```python
df_train = pd.read_csv('D:/Luu/1.ML/1.Linear Regression/train.csv')
df_test = pd.read_csv('D:/Luu/1.ML/1.Linear Regression/test.csv')

# the dropna() method to prevent Nan Values
df_train = df_train.dropna()
df_test = df_test.dropna()
```

```python
# ValueError: Input contains NaN, infinity or a value too large for dtype('float64')
x_train = df_train['x']
y_train = df_train['y']
x_test = df_test['x']
y_test = df_test['y']
```

# Plot raw values

```python
plt.scatter(x_train,y_train, color='red', label='GT')
plt.legend()
plt.show()
```

```python
# Check size of daata
print (x_train.size)
print (y_train.size)
```

# Convert data train and test to array type

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

# Reshape data;
-1 is automatically calculate the rows data; 1 is 1-d columns array

```python
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
```

# Import the regression packages

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Create liner regression object 
clf = LinearRegression(normalize=True)

# Train the model using training data sets
clf.fit(x_train,y_train)

# Make predictions using the testing set
y_pred = clf.predict(x_test)

# Print the cofficients
print("Cofficients: ",r2_score(y_test,y_pred))
print("Mean squared errod: %.2f" % (r2_score(y_test,y_pred)))

# Plot output
plt.figure(figsize =(15,15))
plt.scatter(x_train,y_train, color='red', label='GT')
plt.plot(x_test,y_pred,color='blue', label ='pred')
plt.legend()
plt.show()
```
